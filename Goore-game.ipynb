```python
import random

class Environment:
    def __init__(self, auto_number):
        self.number = auto_number

    def penalty(self, m):
        if m <= 3:
            if random.random() > m * 0.2:
                return True
            else:
                return False
        else:
            if random.random() > 0.6 - (m - 3) * 0.2:
                return True
            else:
                return False

class Tsetlin:
    def __init__(self, n):
        # n is the number of states per action
        self.n = n

        # Initial state selected randomly
        self.state = random.choice([self.n, self.n+1])

    def reward(self):
        if self.state > 1 and self.state <= self.n:
            self.state -= 1
        elif self.state > self.n and self.state < 2 * self.n:
            self.state += 1

    def penalize(self):
        if self.state <= self.n:
            self.state += 1
        elif self.state > self.n:
            self.state -= 1

    def makeDecision(self):
        # No-action
        if self.state <= self.n:
            return 1
        # Yes-action
        else:
            return 2
        

```

```python
# Initialization
env = Environment(5)
la = [Tsetlin(30) for _ in range(env.number)]
action_count = [-1,-1,-1,-1,-1]

print(f"States: {la[0].state, la[1].state, la[2].state, la[3].state, la[4].state}\n")

# Decision loop
for i in range(500):
 
    actions = [la[i].makeDecision() for i in range(len(la))]
    
    for j in range(env.number):
        action_count[j] += actions[j]
    yes_count = sum(action_count)
    
    for j in range(env.number): 
        penalty = env.penalty(yes_count)
        if penalty:
            print(f"Automata {j} : Penalty")
            la[j].penalize()
        else:
            print(f"Automata {j} : Reward")
            la[j].reward()
    
    # Printing decisions and new states
    print(f"Actions (0:No, 1:Yes) : {action_count}")
    print("==============================================\n")  
    print(f"New States: {la[0].state, la[1].state, la[2].state, la[3].state, la[4].state}\n")
    
    action_count = [-1,-1,-1,-1,-1]
```
